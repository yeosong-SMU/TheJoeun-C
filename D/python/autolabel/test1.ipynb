```python
import joblib

svc_label_model = joblib.load("svc_label_model.pkl")
tfidf_vectorizer = joblib.load("tfidf_vectorizer_label.pkl")
label_encoder = joblib.load("label_encoder.pkl")

print("모델 및 인코드 불러오기 완료")

sample_texts = [
    "결제가 이중으로 되었습니다. 빠른 조치 부탁드립니다.",
    "반품 요청을 했는데 언제쯤 환불이 될까요?",
    "카드 결제가 실패했습니다. 확인 바랍니다."
]

sample_tfidf = tfidf_vectorizer.transform(sample_texts)
sample_pred = label_encoder.inverse_transform(svc_label_model.predict(sample_tfidf))

for text, pred in zip(sample_texts, sample_pred):
    print(f"{text} => {pred}")
```
