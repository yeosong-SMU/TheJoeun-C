```python
import pandas as pd
import random

models = [
    ("Avante", 1500), ("Sonata", 2500), ("K5", 2300), 
    ("SM6", 2200), ("Tivoli", 1800), ("Grandeur", 2800), 
    ("Spark", 1200), ("K3", 1600)
]
fuels = ["휘발유", "경유", "LPG", "전기"]
transmissions = ["오토", "수동"]

def generate_car(idx):
    model, base_price = random.choice(models)
    year = random.randint(2010, 2022)
    mileage = random.randint(10000, 200000)
    fuel = random.choice(fuels)
    transmission = random.choice(transmissions)

    #감가 계산
    age_discount = (2025 - year) * 100
    mileage_discount = mileage * 0.01
    fuel_discount = 100 if fuel == "LPG" else 0

    #최저 300만원
    predicted_price = max(int(base_price - age_discount - mileage_discount - fuel_discount), 300)

    return {
        "id": idx, "model": model, "year": year, 
        "mileage": mileage, "fuel": fuel, 
        "transmission": transmission, 
        "predicted_price": predicted_price
    }

data = [generate_car(i+1) for i in range(1000)]
df = pd.DataFrame(data)
df.to_csv("d:/data/car/car_data_sample.csv", index=False)
print('작업 종료')
```
