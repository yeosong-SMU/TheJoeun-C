```python
import pandas as pd
import random
import os

os.makedirs("d:/data/weather", exist_ok=True)

data = []
for _ in range(1000):
    temperature = round(random.uniform(-5, 35), 1)
    humidity = random.randint(30, 100)
    wind_speed = round(random.uniform(0.5, 10), 1)
    rain = random.choice(['yes', 'no'])

    if temperature >= 27:
        clothing = '반팔+우산' if rain == 'yes' else '반팔'
    elif temperature >= 20:
        clothing = '긴팔+우산' if rain == 'yes' else '긴팔'
    elif temperature >= 13:
        clothing = '가디건+우산' if rain == 'yes' else '가디건'
    else:
        clothing = '코트+우산' if rain == 'yes' else '코트'
    
    data.append([temperature, humidity, wind_speed, rain, clothing])

df = pd.DataFrame(data, columns=['temperature', 'humidity', 'wind_speed', 'rain', 'clothing'])
df.to_csv("d:/data/weather/weather_clothing.csv", index=False, encoding='utf-8-sig')

print("완료")
```
