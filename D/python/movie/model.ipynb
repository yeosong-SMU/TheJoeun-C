```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import joblib

# movieId, title, genre, rating_age
df_movies = pd.read_csv('d:/data/movie/movie.csv')

genre_dummies = df_movies['genre'].str.get_dummies(sep='/')

genre_sim_matrix = cosine_similarity(genre_dummies)

df_movies = df_movies.reset_index()
df_movies.rename(columns={'index': 'row_index'}, inplace=True)

movie_id_to_index = dict(zip(df_movies['movieId'], df_movies['row_index']))
index_to_movie_id = dict(zip(df_movies['row_index'], df_movies['movieId']))

joblib.dump({
    'sim_matrix': genre_sim_matrix,
    'df_movies': df_movies,
    'movie_id_to_index': movie_id_to_index,
    'index_to_movie_id': index_to_movie_id
}, 'd:/data/movie/content_recommender.pkl')

print("모델 저장 완료")
```
