```python
import joblib, pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

CSV_PATH = "d:/data/customer/customers.csv"
K = 4
FEATS = ["age", "income", "spend"]

df = pd.read_csv(CSV_PATH)
X = df[FEATS].values

scaler = StandardScaler().fit(X)
Xs = scaler.transform(X)

kmeans = KMeans(n_clusters=K, n_init=10, random_state=42).fit(Xs)

joblib.dump(kmeans, "d:/data/customer/kmeans.joblib")
joblib.dump(scaler, "d:/data/customer/scaler.joblib")

centers = scaler.inverse_transform(kmeans.cluster_centers_)
print(f"[OK] trained k={K}")
print("centers (inverse-scaled): ")
for i, c in enumerate(centers):
    print(f"  #{i}: age={c[0]:.2f}, income={c[1]:.2f}, spend={c[2]:.2f}")
```
