```python
import pandas as pd
import numpy as np
import random
import os

random.seed(42)
np.random.seed(42)

majors = ['컴퓨터공학', '전자공학', '기계공학', '소프트웨어학과', '경영학과']
patterns = {
    '컴퓨터공학' :      {'kor': (80, 5), 'eng': (85, 5), 'math': (90, 5)},
    '전자공학' :        {'kor': (75, 7), 'eng': (80, 7), 'math': (85, 7)},
    '기계공학' :        {'kor': (70, 6), 'eng': (75, 6), 'math': (80, 6)},
    '소프트웨어학과' :  {'kor': (82, 5), 'eng': (88, 4), 'math': (89, 4)},
    '경영학과' :        {'kor': (85, 6), 'eng': (85, 6), 'math': (80, 6)},
}

data = []

for _ in range(1000):
    major = random.choice(majors)
    pattern = patterns[major]

    kor = int(np.clip(np.random.normal(*pattern['kor']), 40, 100))
    eng = int(np.clip(np.random.normal(*pattern['eng']), 40, 100))
    math = int(np.clip(np.random.normal(*pattern['math']), 40, 100))
    total = kor + eng + math

    if total >= 255:
        result = '합격'
    elif total >= 235:
        result = '합격' if random.random() < 0.75 else '불합격'
    elif total >= 215:
        result = '합격' if random.random() < 0.4 else '불합격'
    else:
        result = '불합격'
    
    data.append([kor, eng, math, major, result])

df = pd.DataFrame(data, columns=['kor', 'eng', 'math', 'wish_major', 'result'])

save_path = 'd:/data/admission'
os.makedirs(save_path, exist_ok = True)
df.to_csv(f'{save_path}/admission_data.csv', index=False, encoding='utf-8-sig')

print("완료")
```
